```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
```

# 

```python
dataset = load_boston()

X = dataset.data
y = dataset.target[:,np.newaxis]

print("Total samples in our dataset is: {}".format(X.shape[0]))
```

## Let's find the relationship between the Head Size and Brain weights. 

```python
# initializing our inputs and outputs
X = dataset['Head Size(cm^3)'].values
Y = dataset['Brain Weight(grams)'].values
```

```python
# mean of our inputs and outputs
x_mean = np.mean(X)
y_mean = np.mean(Y)
```

```python
#total number of values

n = len(X)# using the formula to calculate the b1 and b0
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - x_mean) * (Y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2
    
b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean) #printing the coefficient
print(b1, b0)
```

### Now we have our bias coefficient(b) and scale factor(m). In mathematical terms :

#### Brain weights =  325.57342104944223 + 0.26342933948939945 * Head size

### Now we have a linear model.
#### Lets plot it graphically.

```python
#plotting values 
x_max = np.max(X) + 100
x_min = np.min(X) - 100  #calculating line values of x and y
x = np.linspace(x_min, x_max, 1000)
y = b0 + b1 * x  #plotting line 
plt.plot(x, y, color='#00ff00', label='Linear Regression')  #plot the data point
plt.scatter(X, Y, color='#ff0000', label='Data Point')  # x-axis label
plt.xlabel('Head Size (cm^3)')  #y-axis label
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.show()
```
